## 웰컴 키트
https://www.acmicpc.net/problem/30802

```python
import sys
input = sys.stdin.readline

n = int(input())
arr = list(map(int, input().split()))
t, p = map(int, input().split())

t_cnt = 0

# 티셔츠 묶음 계산 (올림)
for t_no in arr:
    if t_no == 0:  # 0개면 안 사도 됨
        continue
    # 올림: (개수 + 묶음크기 - 1) // 묶음크기
    t_cnt += (t_no + t - 1) // t

# 펜 묶음과 낱개
p_set_cnt = n // p  # 정수 나누기!
p_cnt = n % p

print(t_cnt)
print(f"{p_set_cnt} {p_cnt}")
```

## 달팽이는 올라가고 싶다
https://www.acmicpc.net/problem/2869

```python
import sys
input = sys.stdin.readline

A, B, V = map(int, input().split())
cnt = 0

for i in range(1, V + 1):
    if V - A <= 0:
        cnt += 1
        break
    V -= (A-B)
    cnt += 1

print(cnt)
```

```python
# 정답 코드
import sys
input = sys.stdin.readline

A, B, V = map(int, input().split())

# 마지막 날 제외한 높이
need = V - A

# 하루에 올라가는 순 높이
per_day = A - B

# 올림 계산
if need <= 0:  # 첫날에 바로 도착
    print(1)
else:
    days = (need + per_day - 1) // per_day + 1  # 마지막 날 추가
    print(days)
```

```python
# 정답 코드, 메모리 관리를 위한 계수 정리임
import sys
input = sys.stdin.readline

n = int(input())

# 1~10,000 범위의 개수만 카운트
count = [0] * 10001  # 인덱스 0은 안 씀

# 입력받으면서 바로 카운트
for _ in range(n):
    num = int(input())
    count[num] += 1

# 카운트한 만큼 출력
for i in range(1, 10001):
    if count[i] > 0:
        for _ in range(count[i]):
            print(i)
```

```python
isbn = input().strip()

# * 위치와 가중치 찾기
total = 0
star_weight = 0

for i in range(13):
    if isbn[i] == '*':
        star_weight = 1 if i % 2 == 0 else 3
    else:
        weight = 1 if i % 2 == 0 else 3
        total += int(isbn[i]) * weight

# 0~9 넣어보기
for num in range(10):
    if (total + num * star_weight) % 10 == 0:
        print(num)
        break
```

```python
import sys
input = sys.stdin.readline

# 3줄 입력
a = input().strip()
b = input().strip()
c = input().strip()

# 숫자 찾기
next_num = 0

if a.isdigit():  # 첫 줄이 숫자면
    next_num = int(a) + 3
elif b.isdigit():  # 둘째 줄이 숫자면
    next_num = int(b) + 2
elif c.isdigit():  # 셋째 줄이 숫자면
    next_num = int(c) + 1

# FizzBuzz 판정
if next_num % 15 == 0:  # 3과 5의 배수
    print("FizzBuzz")
elif next_num % 3 == 0:  # 3의 배수만
    print("Fizz")
elif next_num % 5 == 0:  # 5의 배수만
    print("Buzz")
else:
    print(next_num)
```

```python
import sys
input = sys.stdin.readline

n = int(input())
people = []

# 입력 받기
for _ in range(n):
    w, h = map(int, input().split())
    people.append((w, h))

# 각 사람의 등수 계산
ranks = []
for i in range(n):
    rank = 1  # 기본 1등
    for j in range(n):
        if i == j:  # 자기 자신은 제외
            continue
        # 둘 다 작으면 등수 +1
        if people[i][0] < people[j][0] and people[i][1] < people[j][1]:
            rank += 1
    ranks.append(rank)

# 출력
print(' '.join(map(str, ranks)))
```
