# 1번 사이즈 추천

## 입력
첫째 줄:
B, Q  (브랜드 개수, 고객 수) - 쉼표로 구분 <br/>
다음 B개 브랜드 정보:<br/>
브랜드명, 상품개수N<br/>
사이즈명, 키min, 키max, 가슴min, 가슴max, 허리min, 허리max<br/>
사이즈명, 키min, 키max, 가슴min, 가슴max, 허리min, 허리max<br/>
...N개<br/><br/>
다음 Q개 고객 정보:<br/>
원하는_브랜드명, 고객_키, 고객_가슴, 고객_허리<br/><br/>

제약사항:<br/>
- 모든 값은 쉼표(, )로 구분
-  치수는 6개 (키, 가슴, 허리 각각 min, max)
- 모든 수는 정수

## 출력
Q개의 줄에 걸쳐 각 고객에 대해:
브랜드명, 결과
결과 우선순위:

브랜드 없음 → unknown
맞는 사이즈 있음 → 사이즈명 (여러 개면 입력 순서상 첫 번째)
맞는 사이즈 없을 때:

고객의 모든 치수가 모든 사이즈의 최솟값보다 작음 → DOWN
고객의 모든 치수가 모든 사이즈의 최댓값보다 큼 → UP
그 외 (일부는 작고 일부는 큼) → mismatch

## 조건 상세
### 맞는 사이즈 조건
키min ≤ 고객_키 ≤ 키max AND
가슴min ≤ 고객_가슴 ≤ 가슴max AND
허리min ≤ 고객_허리 ≤ 허리max
모든 조건을 동시에 만족해야 함
### DOWN 조건
고객_키 < 모든 사이즈의 키min AND
고객_가슴 < 모든 사이즈의 가슴min AND
고객_허리 < 모든 사이즈의 허리min
### UP 조건
고객_키 > 모든 사이즈의 키max AND
고객_가슴 > 모든 사이즈의 가슴max AND
고객_허리 > 모든 사이즈의 허리max
### mismatch 조건
DOWN도 UP도 아닌 경우
예: 키는 맞는데 허리가 큼, 또는 가슴은 작은데 키는 큼

## 예제
### 입력
2, 4
Nike, 3
S, 160, 170, 80, 90, 70, 80
M, 170, 180, 85, 95, 75, 85
L, 180, 190, 90, 100, 80, 90
Adidas, 2
M, 165, 175, 82, 92, 72, 82
L, 175, 185, 87, 97, 77, 87
Nike, 165, 85, 75
Nike, 155, 75, 65
Nike, 195, 105, 95
Adidas, 182, 95, 85
### 출력
Nike, S
Nike, DOWN
Nike, UP
Adidas, L

```python
import sys
input = sys.stdin.readline

B, Q = map(int, input().split(', '))

# 브랜드별 상품 정보 저장
brands = {}
for _ in range(B):
    line = input().strip().split(', ')
    brand_name = line[0]
    N = int(line[1])
    
    products = []
    for _ in range(N):
        specs = input().strip().split(', ')
        size_name = specs[0]
        measures = list(map(int, specs[1:7]))
        products.append((size_name, measures))
    
    brands[brand_name] = products

# 고객 쿼리 처리
for _ in range(Q):
    line = input().strip().split(', ')
    want_brand = line[0]
    H, C, W = map(int, line[1:4])
    
    # 브랜드 없음
    if want_brand not in brands:
        print(f"{want_brand}, unknown")
        continue
    
    # 맞는 사이즈 찾기
    found_size = None
    for size_name, measures in brands[want_brand]:
        h_min, h_max, c_min, c_max, w_min, w_max = measures
        
        if (h_min <= H <= h_max and 
            c_min <= C <= c_max and 
            w_min <= W <= w_max):
            found_size = size_name
            break
    
    # 맞는 사이즈 있음
    if found_size:
        print(f"{want_brand}, {found_size}")
        continue
    
    # 맞는 사이즈 없음 - DOWN/UP/mismatch 판단
    all_too_big = True
    all_too_small = True
    
    for size_name, measures in brands[want_brand]:
        h_min, h_max, c_min, c_max, w_min, w_max = measures
        
        if H >= h_min or C >= c_min or W >= w_min:
            all_too_big = False
        
        if H <= h_max or C <= c_max or W <= w_max:
            all_too_small = False
    
    if all_too_big:
        print(f"{want_brand}, DOWN")
    elif all_too_small:
        print(f"{want_brand}, UP")
    else:
        print(f"{want_brand}, mismatch")
```
