```python
### 다른 답변
N = int(input())
data = [list(map(int, input().strip().split())) for _ in range(N)]

x = sorted([(i[0]) for i in data])
y = sorted([(i[1]) for i in data])

median_x = x[N // 2]
median_y = y[N // 2]

cnt = 0

for i in range(N):
	cnt += abs(data[i][0] - median_x)
	cnt += abs(data[i][1] - median_y)

print(cnt)
```

```python
### LLM 정답
import sys
input = sys.stdin.readline

n = int(input())
x_list = []
y_list = []

# 입력 받기
for _ in range(n):
    x, y = map(int, input().split())
    x_list.append(x)
    y_list.append(y)

# 정렬
x_list.sort()
y_list.sort()

# 중앙값 찾기
median_x = x_list[n // 2]
median_y = y_list[n // 2]

# 거리 합 계산
total = 0
for i in range(n):
    total += abs(x_list[i] - median_x)
    total += abs(y_list[i] - median_y)

print(total)
```
